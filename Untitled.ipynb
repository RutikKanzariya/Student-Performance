### data Cleaning and target variable exploration

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['average_score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
```

```python
df['average_score'] = df['average_score'].round(2)
```

```python
df.head()
```

```python
for col in num_col:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)


    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR


    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    print(f"{col} : {len(outliers)} Outliers.")
```

```python
df['result'] = df['average_score'].apply(
    lambda x : 1 if x >= 60 else 0
)
```

```python
df.head()
```

```python
df['result'].value_counts()
```

```python
sns.countplot(x='result',data=df)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import *
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_encoded,y,test_size=0.2,train_size=0.8,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
scaler = StandardScaler()
```

```python
print("X_train:", x_train.shape)
print("X_test :", x_test.shape)

print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

### Save the Scaler 

```python
import joblib
```

```python
joblib.dump(scaler,'scaler.pkl')
```

### Train the Models

```python
models = {
    "Logistic Regression" : LogisticRegression(),
    "K-Nearest Neighbour" : KNeighborsClassifier(n_neighbors=5),
    "Gaussian " : GaussianNB(),
    "Decision tree" : DecisionTreeClassifier(max_depth=5,random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    'SVC' : SVC(random_state=42)
}
```

```python
result = []
```

```python
for name,model in models.items():
    model.fit(x_train_scaled,y_train)
    y_pred = model.predict(x_test_scaled)
    accuracy = accuracy_score(y_test,y_pred)
    matrix = confusion_matrix(y_test,y_pred)
    report = classification_report(y_test,y_pred)
    print(f"Classification Report for {name}: \n",report)
    result.append({
        "Model" : name,
        "Accuracy" : accuracy
    })
```

```python
result
```

```python
# Model Mis-leading because Mostly Student Pass 
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
```

```python
x_resampled ,y_resampled = smote.fit_resample(
    x_train_scaled,
    y_train
)
```

```python
result_smote = []
```

```python
for name,model in models.items():
    model.fit(x_resampled,y_resampled)
    y_pred = model.predict(x_test_scaled)
    accuracy = accuracy_score(y_test,y_pred)
    matrix = confusion_matrix(y_test,y_pred)
    report = classification_report(y_test,y_pred)
    print(f"Classification Report for {name}: \n",report)
    # print(f"\n\n{accuracy}\n\n")
    result_smote.append({
        "Model" : name,
        "Accuracy" : accuracy
    })
```

```python
result_smote
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(
    LogisticRegression(),
    x_resampled,
    y_resampled,
    cv=5,
    scoring='f1_macro'
)
```

```python
print(scores)
print("Mean:", scores.mean())
print("Std:", scores.std())
```

### HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "solver": ["liblinear", "lbfgs"]
}

grid = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid,
    cv=5,
    scoring="f1_macro",
    n_jobs=-1,
    return_train_score=False
)

grid.fit(x_resampled,y_resampled)
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
res = pd.DataFrame(grid.cv_results_)
```

### Train the Final Model

```python
final_model = LogisticRegression(C=0.01,solver='liblinear',max_iter=1000)

final_model.fit(x_resampled,y_resampled)
```

```python
y_pred = final_model.predict(x_test_scaled)

print(classification_report(y_test, y_pred))
```
